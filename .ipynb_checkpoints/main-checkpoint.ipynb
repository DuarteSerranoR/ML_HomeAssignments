## 1.1. Dataset preparation
### 1.1.1. Shuffle and Split into training and IVS
- `train_df` -- training set for model training/selection
- `ivs_df` -- independet validation set (IVS) for the final benchmark

```python
df = df.sample(frac=1)
train_size = int(0.8 * n)
train_df = df.iloc[:train_size]
ivs_df = df.iloc[train_size:]
```

```python
#train_x = np.array(train_df[['Jitter(%)', 'Shimmer', 'NHR', 'HNR', 'RPDE', 'DFA', 'PPE']])
train_x = np.array(train_df[['age', 'sex', 'HNR', 'RPDE', 'DFA', 'PPE']])
train_y = np.array(train_df["motor_UPDRS"])

#ivs_x = np.array(ivs_df[['Jitter(%)', 'Shimmer', 'NHR', 'HNR', 'RPDE', 'DFA', 'PPE']])
ivs_x = np.array(ivs_df[['age', 'sex', 'HNR', 'RPDE', 'DFA', 'PPE']])
ivs_y = np.array(ivs_df["motor_UPDRS"])
```

### 1.1.2. Scale data

```python
x_mean  = np.mean(train_x, axis=0)
x_std   = np.std(train_x, axis=0)
train_x = (train_x - x_mean) / x_std
y_mean  = np.mean(train_y, axis=0)
y_std  = np.std(train_y, axis=0)
train_y = (train_y - y_mean) / y_std
ivs_x_scaled = (ivs_x - np.mean(ivs_x, axis=0)) / np.std(ivs_x, axis=0)
ivs_y_scaled   = (ivs_y - np.mean(ivs_y, axis=0)) / np.std(ivs_y, axis=0)
```

## 1.2. Model Training and Selection

### 1.2.1. Validation techniques (shared for regression and classification)

```python
def simple_cross_validation(*, model, x, y, train_partition=(0, 0.8), metrics={"mae": mean_absolute_error}):
    n, m = x.shape
    
    start = int(train_partition[0] * n)
    end = int(train_partition[1] * n)
    
    x_train = x[start:end, :]
    y_train = y[start:end]
    x_test = np.concatenate((x[:start, :], x[end:, :]))
    y_test = np.concatenate((y[:start], y[end:]))
    
    #print("SCV")
    #print(x_train)
    #print(y_train)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    
    return {metric_name: metric_f(y_test, y_pred) for (metric_name, metric_f) in metrics.items()}


def kfold(*, model, x, y, k=5, metrics={"mae": mean_absolute_error}):
    n, m = x.shape
    results = []
    
    for i in range(k):
        start = i * (1/k)
        end = start + (1/k)
        if (i + 1) == k:
            end = 1.0
            
        result = simple_cross_validation(model=model, x=x, y=y, train_partition=(start, end), metrics=metrics)
        
        results.append(result)
    
    return results
```

### 1.2.2. Model Selection

```python
def tree_regressor_selection(x: np.ndarray, y: np.ndarray, metrics):
    # Hyperparameters
    criterions = ["squared_error"]
    splitters = ["best", "random"]
    max_depths = [3, 4, 6, 12, 20, None]
    hyper_parameters = product(criterions, splitters, max_depths)
    
    results = {}
    
    for p in hyper_parameters:
        criterion, splitter, max_depth = p
        
        model = tree.DecisionTreeRegressor(criterion=criterion,
                                           splitter=splitter,
                                           max_depth=max_depth)
        
        
        # Use either simple_cross_validation or kfold below
        #result = simple_cross_validation(model=model, x=x, y=y, metrics={"mae": mean_absolute_error, "rve": explained_variance_score})
        result = kfold(model=model, x=x, y=y, k=5, metrics=metrics)
        
        avg_metrics = {f"avg_{metric}": sum([r[metric] for r in result]) / len(result) for metric in metrics.keys()}
        
        model_name = "tree-" + "-".join([str(param) for param in p])
        results[model_name] = {"model": model, "evaluation": result, "avg_evaluation": avg_metrics}
        
    # TODO: select the best model and return it
    return results


def linear_regression_selection(x: np.ndarray, y: np.ndarray, metrics):
    results = {}

    model = linear_model.LinearRegression()
    result = kfold(model=model, x=x, y=y, k=5, metrics=metrics)
    avg_metrics = {f"avg_{metric}": sum([r[metric] for r in result]) / len(result) for metric in metrics.keys()}
    result.append(avg_metrics)
    results[f"linearreg"] = {"model": model, "evaluation": result, "avg_evaluation": avg_metrics}

    return results


def ridge_selection(x: np.ndarray, y: np.ndarray, metrics):
    results = {}
    
    alphas = [*[10**p for p in range(-5,0)], *[]]
    for alpha in alphas:
        model = linear_model.Ridge(alpha=alpha)
        result = kfold(model=model, x=x, y=y, k=5, metrics=metrics)
        avg_metrics = {f"avg_{metric}": sum([r[metric] for r in result]) / len(result) for metric in metrics.keys()}
        result.append(avg_metrics)
        results[f"ridge-{alpha}"] = {"model": model, "evaluation": result, "avg_evaluation": avg_metrics}
        
    return results


def lasso_selection(x: np.ndarray, y: np.ndarray, metrics):
    results = {}
    
    alphas = [*[10**p for p in range(-5,0)], *[]]
    for alpha in alphas:
        model = linear_model.Lasso(alpha=alpha)
        result = kfold(model=model, x=x, y=y, k=5, metrics=metrics)
        avg_metrics = {f"avg_{metric}": sum([r[metric] for r in result]) / len(result) for metric in metrics.keys()}
        result.append(avg_metrics)
        results[f"lasso-{alpha}"] = {"model": model, "evaluation": result, "avg_evaluation": avg_metrics}
        
    return results

        
def model_selection(x: np.ndarray, y: np.ndarray):
    results = {}
    metrics = {"mae": mean_absolute_error, "rve": explained_variance_score, "r2": r2_score}
    results = { **results, **tree_regressor_selection(x, y, metrics) }
    results = { **results, **linear_regression_selection(x, y, metrics) }
    results = { **results, **lasso_selection(x, y, metrics) }
    results = { **results, **ridge_selection(x, y, metrics) }
    
    best_model_name = max(results, key=lambda model: results[model]["avg_evaluation"]["avg_rve"])
    best_model = {**results[best_model_name], **{"name": best_model_name}}
        
    return best_model, results

best_regressor, regressor_results = model_selection(train_x, train_y)
print(f"Best Model: {best_regressor['name']}")
print(f"Best Model Data: {best_regressor['avg_evaluation']}")
# print(json.dumps({"-".join([str(p) for p in k]): v["avg_evaluation"] for (k, v) in results.items()}, indent=4))
```

## 1.3. Validate against IVS

```python
def validate_agains_ivs(*, model, ivs_x, ivs_y, metrics={"mae": mean_absolute_error, "rve": explained_variance_score}):
    y_pred = model.predict(ivs_x)
    
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.axline((0, 0), slope=1, color="black")
    plt.plot(y_pred, ivs_y, ".")
    
    return {metric_name: metric_f(ivs_y, y_pred) for (metric_name, metric_f) in metrics.items()}

    
performance_on_ivs = validate_agains_ivs(model=best_regressor["model"], ivs_x=ivs_x_scaled, ivs_y=ivs_y_scaled)
print(f"Performance on IVS: {performance_on_ivs}")
print(f"MAE rescaled: {performance_on_ivs['mae'] * y_std} ")
```
